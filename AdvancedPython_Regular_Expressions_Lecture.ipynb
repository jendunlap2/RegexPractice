### Sets <br>
<p>The following cells will allow you to use regular expressions to search for certain values within a range such as numbers 1 through 4.</p>

##### [a-z] or [A-Z] - any lowercase/uppercase letters from a to z<br/>[^2] - anything that's not 2

##### Integer Ranges

```python
int_pattern = re.compile('[0-5]')

random_num = '3278194054371289'

low_nums = int_pattern.findall(random_num)
print(low_nums)
```

```python
int_pat = re.compile('[5-9][0-5]')
nums = int_pat.findall(random_num)
print(nums)
```

```python
new_num = '951'
int_pat.findall(new_num)
```

##### Character Ranges

```python
char_pattern = re.compile('[A-Z][a-z]') # Uppercase Letter followed by lowercase letter

found = char_pattern.findall('Hello There Mr. Anderson')
print(found)
```

```python
another_pattern = re.compile('[lksdf]')

another_pattern.findall('hello this is a test string to find letters')
```

### Counting Occurences

##### {x} - something that occurs {num_of_times}

```python
my_address = '555 Circle Drive, Chicago, IL 60663'

zip_code_pattern = re.compile('[0-9]{5}')

zip_code_pattern.findall(my_address)
```

```python
char_pat_count = re.compile('[A-Z][a-z]{2}[lr]')

found_chars = char_pat_count.findall('Hello Mr. Anderson')
print(found_chars)
```

##### {x, x} - something that occurs between x and y times

```python
my_address = '555 Circle Drive, Chicago, IL 60663-4583'

three_to_five = re.compile('[0-9]{3,5}') # numbers 0-9 that occur BETWEEN 3 and five times in a row

three_to_five.findall(my_address)
```

##### ? - something that occurs 0 or 1 time

```python
pre_pattern = re.compile('Mr?s?')

found_pre = pre_pattern.findall('Hello Mr. and Mrs. Anderson. How are you? Have you seen Ms. Smith?')

print(found_pre)
```

##### * - something that occurs at least 0 times

```python
pattern_m = re.compile('M*s')

found_m = pattern_m.findall('MMMs name is Ms. Smith. This i sMsssss MM.')

print(found_m)
```

##### + - something that occurs at least once

```python
pattern_m = re.compile('M+s')

found_m = pattern_m.findall('MMMs name is Ms. Smith. This is Msssss MM.')

print(found_m)
```

### Escaping Characters

##### \w - look for any Unicode character<br/>\W - look for anything that isnt a Unicode character

```python
sentence = "This is a sentence. with an, exclamation mark at the end! 123-456"

letter_pattern = re.compile('\w')

letters = letter_pattern.findall(sentence)

print(letters)
```

```python
non_letter_pattern = re.compile('\W')
non_letters = non_letter_pattern.findall(sentence)
print(non_letters)
```

```python
word_pattern = re.compile('\w+')
words = word_pattern.findall(sentence)

print(words)
```

### Grouping

```python
my_string_again = "Max Smith, aaron rodgers, Sam Darnold,LeBron James, Micheal Jordan, Kevin Durant, Patrick McCormick"

# Group of names Regular Expression Compiler

name_pattern = re.compile('([A-Z][a-zA-Z]+) ([A-Z][a-zA-Z]+)')

found_names = name_pattern.findall(my_string_again)
print(found_names)

#Loop over the list to get the tuples
for name in found_names:
    print(name[1])
```

```python
for name in my_string_again.split(','):
    match = name_pattern.search(name)
    if match:
        print(match.group())
    else:
        print(f"{name} is a bad guy.")
```

##### In-class Exercise 2:

Write a function using regular expressions to find the domain name in the given email addresses (and return None for the invalid email addresses)<br><b>HINT: Use '|' for either or</b>

```python
my_emails = ["jordanw@codingtemple.orgcom", "pocohontas1776@gmail.com", "helloworld@aol..com",
             "yourfavoriteband@g6.org", "@codingtemple.com"]

# You can also use the $ at the end of your compile expression -- this stops the search

#Expected output:
#None
#pocohontas1776@gmail.com
#None
#yourfavoriteband@g6.org
#None

def emails(email_list):
    for email in email_list:
        is_valid = re.compile('\w+@\w+.(org|com)$')
        is_match = is_valid.match(email)
        
        if is_match:
            print(email)
        else:
            print("NONE")
        
emails(my_emails)
```
